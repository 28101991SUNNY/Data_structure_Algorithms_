```python
A = [1,6,1,1,2,6,6,6,6,6,1]

def majority_element_N_2(A):
    n = len(A)

    ME = None
    f = 0
    
    for i in range(n):

        if f == 0:
            ME = A[i]
            f = 1

        elif ME == A[i]:
            f += 1
        else:
            f -= 1
            if f == 0:
                ME = None
    c = 0
    for e in A:
        if e == ME:
            c = c+1
            if c > n/2:
                return e
majority_element_N_2(A)
```

# You're given a read-only array of N integers. Find out if any integer occurs more than N/3 times in the array in linear time and constant additional space.
If so, return the integer. If not, return -1.

If there are multiple solutions, return any one.

```python
def repeatedNumber(A):
    n = len(A)

    ME1 = None
    f1 = 0

    ME2 = None
    f2 = 0

    for i in range(n):

        if ME1 == A[i]:
            f1 += 1
        elif ME2 == A[i]:
            f2 += 1

        elif f1 == 0:
            f1 += 1
            ME1 = A[i]
        elif f2 == 0:
            f2 += 1
            ME2 = A[i]
        else:
            f1 -= 1
            f2 -= 1

    freq1 = 0
    freq2 = 0

    for i in range(n):
        if A[i]== ME1:
            freq1 += 1
        elif A[i] == ME2:
            freq2 += 1
    if freq1 > n/3:
        return ME1
    if freq2 > n/3:
        return ME2
    return -1
```

```python
def removing_subarray(A,B):
    n = len(A)

    flag = True
    for i in range(1,n):
        if abs(A[i]-A[i-1]) > B :
             flag = False
    if flag == True:
        return 0
    
    flag = True
    for i in range(1,n):
        if abs(A[i]-A[i-1]) <= B :
             flag = False
    if flag == True:
        return (len(A)-1)
    
    s = 0
    e = n-1

    for i in range(0,n-1):
        if abs(A[i]-A[i+1]) <= B:
            s +=1
        else:
            break
    for i in range(n-2,1,-1):
        if abs(A[i]-A[i-1]) <= B:
            e -= 1
        else:
            break 
    
    
    e = e-1
    print("s:",s,"\ne:",e)
    minlen = float('inf')
    for i in range(0,s+1):
        for j in range(e,n):
            
            length = j-i+1

            if i != 0 and j != n-1 :
                if abs(A[i-1]-A[j+1]) <= B:
                    minlen = min(minlen,length)
            else:
                minlen = min(minlen,length)

    return minlen
```

```python

n = 8

B = [[3,8,0],
    [2,5,3],
    [1,0,0],
    [1,0,0],
    [3,8,0],
    [3,2,0],
    [3,8,0],
    [2,4,4],
    [3,2,0],
     [2,7,2],
     [2,1,8]]
A = [[0]*n for _ in range(n)]



answer = []
for i in range(len(B)):
    sum = 0
    if (B[i][0]) == 1:
        
        
        for x in range(n):
            for y in range(n):
                sum += A[x][y]
        answer.append(sum)
        
    elif (B[i][0]) == 2:
        x = B[i][1]-1
        y = B[i][2]-1
        if A[x][y] == 1:
            A[x][y] = 0
        else:
            A[x][y] = 1
            
                         # invert all 
    elif (B[i][0]) == 3:    # for B[i][0] = 3
        r = B[i][1]-1

        for c in range(n):
            if A[r][c] == 1:
                A[r][c] = 0
            else:
                A[r][c] = 1

print(answer)
```
