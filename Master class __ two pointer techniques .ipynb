# given a sorted array , check it there exist a pair (i,j) such tht i != j and A[i] + A[j] == k ,

```python
def sumsubarrayk(A,k):            # total iterations max n-1 wost case senatios TC :O(n)
                                    # only if the gievn array is soted this will work
    n = len(A)

    pointer1 = 0
    pointer2 = n-1

    while pointer1 < pointer2:

        if A[pointer1]+A[pointer2] > k:
            pointer2 -= 1
        elif A[pointer1]+ A[pointer2] < k:
            pointer1 += 1
        else:
            return A[pointer1],A[pointer2]
    return "NO pair found"
        
A = [-3,0,1,3,6,8,11,14,18,25]
k = 11
sumsubarrayk(A,k)
```

# Q : given 3 sorted arrays A,B,C
# find i , j , k such tht 

## max(A[ i ],B,[ j ],C[ k ]) - min(A[ i ],B,[ j ],C[ k ])  is minimum

```python

def asdassd(A,B,C):
    
    p1 = 0
    p2 = 0
    p3 = 0
    answer = float('inf')
    differences  = []
    indexes = []
    while p1 < len(A) and p2 < len(B) and p3 < len(C):
        maxval = max(A[p1],B[p2],C[p3])
        minval = min(A[p1],B[p2],C[p3])
        answer = min(answer,(maxval - minval))
        differences.append(maxval - minval)
        indexes.append([p1,p2,p3])
        
        if A[p1] < B[p2] and A[p1] < C[p3]:
            p1 += 1
        elif B[p2] < C[p3]:
            p2 += 1
        else:
            p3 += 1
    return answer, differences,indexes

A = [3,14,16,20,29,40,55]
B = [ -6, 23,24,30,35,50]
C  = [-15,15,26,29,31,39,42]
asdassd(A,B,C)
```
